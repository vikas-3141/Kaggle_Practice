```python
import pandas as pd
import numpy as np
```

#### downloading the Minimim Wage Data.csv file. <a href="https://www.kaggle.com/datasets/lislejoem/us-minimum-wage-by-state-from-1968-to-2017">Minimum Wage Data<a/>

```python
df=pd.read_csv("Minimum Wage Data.csv",encoding="latin")
```

```python
df.head()
```

```python
gb=df.groupby("State")
gb.get_group("Alabama").set_index("Year").head()
```

```python
min_wage = pd.DataFrame()
for name,group in df.groupby("State"):
    if min_wage.empty:
        min_wage=group.set_index("Year")[["State.Minimum.Wage.2020.Dollars"]].rename(columns={"State.Minimum.Wage.2020.Dollars":name})
    else:
        min_wage=min_wage.join(group.set_index("Year")[["State.Minimum.Wage.2020.Dollars"]].rename(columns={"State.Minimum.Wage.2020.Dollars":name}))
min_wage.head()
```
