```python
import pandas as pd 
df=pd.read_csv("avocado.csv")
df=df.copy()[df['type']=="organic"]
dt['Date']=pd.to_datetime(df['Date'])

dt.sort_values(by="Date",ascending=True,inplace=True)
dt.head()
graph_dt = pd.DataFrame()
for region in l1[:16]:
    print(region)
    region_dt=dt.copy()[dt['region'] == region]
    region_dt.set_index('Date',inplace=True)
    region_dt.sort_index(inplace=True)
    region_dt['Mean']=region_dt['AveragePrice'].rolling(25).mean()
    
    if graph_dt.empty:
        graph_dt=region_dt[['Mean']]
    else:
        graph_dt=graph_dt.join(region_dt['Mean'], rsuffix=f'_{region}')
```
